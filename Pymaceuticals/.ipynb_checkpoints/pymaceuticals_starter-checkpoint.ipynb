```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
raw_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="inner" ) #'raw_df' includes duplicates

# Display the data table for preview
raw_df
```

```python
# Checking the number of mice.
raw_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_duplicates = raw_df.duplicated(subset=["Mouse ID","Timepoint"], keep=False)
dup_mouse_id = raw_df[mouse_duplicates]["Mouse ID"]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
raw_df[raw_df["Mouse ID"].isin(dup_mouse_id)]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
df = raw_df[~raw_df["Mouse ID"].isin(dup_mouse_id)]
df
```

```python
# Checking the number of mice in the clean DataFrame.
df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regi_group = df.groupby(by="Drug Regimen")
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
regi_summary = pd.DataFrame()
regi_summary["Mean"] = regi_group["Tumor Volume (mm3)"].mean()
regi_summary["Median"] = regi_group["Tumor Volume (mm3)"].median()
regi_summary["Variance"] = regi_group["Tumor Volume (mm3)"].var()
regi_summary["Std"] = regi_group["Tumor Volume (mm3)"].std()
regi_summary["SEM"] = regi_group["Tumor Volume (mm3)"].sem()
regi_summary
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
regi_group["Tumor Volume (mm3)"].agg(func=['mean', 'median', 'var','std', 'sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens 
# Put treatments into a list for for loop (and later for plot labels)
some_drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"] # hey check out these drugs
# Start by getting the last (greatest) timepoint for each mouse
mouse_group = df[df["Drug Regimen"].isin(some_drugs)].groupby("Mouse ID")
last_measures = mouse_group["Timepoint"].max()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_tp_df = pd.merge(df, last_measures, on=["Mouse ID","Timepoint"], how="inner")

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = [] 
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
fig, axs = plt.subplots(1,len(some_drugs), constrained_layout=True)

sub_pos = 0
for drug in some_drugs:
    #Get Subset
    this_df = last_tp_df[last_tp_df["Drug Regimen"]==drug]
    #Box Plot
    #plt.figure()
    axs[sub_pos].boxplot(this_df["Tumor Volume (mm3)"], flierprops={'markerfacecolor':'r', 'marker':'*'})
    axs[sub_pos].set_title(drug)
    #plt.show()
    #Print drug
    print("-------------")
    print(drug)
    #Get Quartiles
    Q1 = this_df["Tumor Volume (mm3)"].quantile(0.25)
    Q2 = this_df["Tumor Volume (mm3)"].quantile(0.75)
    IQR = Q2-Q1
    #Print Quartiles and IQR
    print(f"Q1: {Q1:.2f} cubic mm")
    print(f"Q2: {Q2:.2f} cubic mm")
    print(f"IQR: {IQR:.2f} cubic mm")
    # Determine outliers using upper and lower bounds
    upper_outliers = this_df[this_df["Tumor Volume (mm3)"]>(Q2+1.5*IQR)]
    [print(f"Upper Outlier: {x:.2f} cubic mm") for x in upper_outliers["Tumor Volume (mm3)"]]
    lower_outliers = this_df[this_df["Tumor Volume (mm3)"]<(Q1-1.5*IQR)]
    [print(f"Lower Outlier: {x:.2f} cubic mm") for x in lower_outliers["Tumor Volume (mm3)"]]
    print("-------------")
    sub_pos += 1
fig.suptitle("Tumor Volume Distribution", fontsize=16)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capo_mice = df[df["Drug Regimen"]=="Capomulin"] # Get the mice
capo_mouse_id = capo_mice["Mouse ID"].iloc[6] # Get the first ID
capo_mouse_data = df[df["Mouse ID"]==capo_mouse_id]
capo_mouse_data.plot(x="Timepoint", y="Tumor Volume (mm3)")
plt.ylabel("Volume (cubic mm)")
plt.xlabel("Days of Treatment")
plt.title(f"Change in Tumor Volume over time for {capo_mouse_id}")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capo_group = capo_mice.groupby("Mouse ID")
capo_weight = capo_group["Weight (g)"].mean()
capo_volume = capo_group["Tumor Volume (mm3)"].mean()
lin_reg = st.linregress(capo_weight, capo_volume)
plt.figure()
plt.scatter(capo_weight, capo_volume, c="b")
plt.title("Weight vs Average Tumor Mass for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Mass (mm3)")
plt.plot(capo_weight, capo_weight*lin_reg.slope+lin_reg.intercept, c="r")
plt.annotate(f"r = {lin_reg.rvalue:.3f}", (22,34.5), c='r')
plt.annotate(f"y = {lin_reg.slope:.3f}*x + {lin_reg.intercept:.3f}", (22,35.5), c='r')
plt.show()
```
